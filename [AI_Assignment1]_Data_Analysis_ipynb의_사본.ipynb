## Graphic visualization of skewness

**[TODO]** Select the Attributes that skewness value is the largest, smallest and the closet to zero.
* For zero skewness attribue, you can specify attribute. e.g., `df['RPM']`

```python
import pandas as pd

# CSV 파일 읽기
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

# 숫자 데이터가 포함된 열 선택
numeric_columns = df.select_dtypes(include=['number']).columns

# 숫자 속성의 왜도 측정
skewness = df[numeric_columns].skew()

# 왜도 값이 가장 큰 속성
highSkewAttribute = df[skewness.idxmax()]

# 왜도 값이 가장 작은 속성
lowSkewAttribute = df[skewness.idxmin()]

# 왜도 값이 0에 가장 가까운 속성
zeroSkewAttribute = df['Length']  # 'Length'는 예시로 주어진 속성

# 결과 출력
print("High skew attribute:\n", highSkewAttribute)
print("\nLow skew attribute:\n", lowSkewAttribute)
print("\nZero skew attribute:\n", zeroSkewAttribute)
```

Draw plots. See how the distribution depends on the Skewness

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# CSV 파일 읽기
df = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/Cars93_miss.csv')

# 숫자 데이터가 포함된 열 선택
numeric_columns = df.select_dtypes(include=['number']).columns

# 숫자 속성의 왜도 측정
skewness = df[numeric_columns].skew()

# 왜도 값이 가장 큰 속성
highSkewAttribute = skewness.idxmax()

# 왜도 값이 가장 작은 속성
lowSkewAttribute = skewness.idxmin()

# 왜도 값이 0에 가장 가까운 속성
zeroSkewAttribute = 'Length'  # 예시로 주어진 속성

# 플롯 그리기
fig, axes = plt.subplots(3, 1, figsize=(10, 15))

# 문자열 설정
strInPlot = "Skewness: %f"

# Plot distplot (highSkew)
ax = axes[0]
sns.histplot(data=df, x=highSkewAttribute, color="skyblue", ax=ax, kde=True)
ax.text(x=0.97, y=0.97, transform=ax.transAxes, s=strInPlot % skewness[highSkewAttribute],
        fontweight='demibold', fontsize=10, verticalalignment='top', horizontalalignment='right')
ax.set_title(f'Distribution of {highSkewAttribute}')

# Plot distplot (zeroSkew)
ax = axes[1]
sns.histplot(data=df, x=zeroSkewAttribute, color="khaki", ax=ax, kde=True)
ax.text(x=0.97, y=0.97, transform=ax.transAxes, s=strInPlot % df[zeroSkewAttribute].skew(),
        fontweight='demibold', fontsize=10, verticalalignment='top', horizontalalignment='right')
ax.set_title(f'Distribution of {zeroSkewAttribute}')

# Plot distplot (lowSkew)
ax = axes[2]
sns.histplot(data=df, x=lowSkewAttribute, color="salmon", ax=ax, kde=True)
ax.text(x=0.97, y=0.97, transform=ax.transAxes, s=strInPlot % skewness[lowSkewAttribute],
        fontweight='demibold', fontsize=10, verticalalignment='top', horizontalalignment='right')
ax.set_title(f'Distribution of {lowSkewAttribute}')

plt.tight_layout()
plt.show()
```
